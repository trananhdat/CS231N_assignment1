k - Nearest Neighbor (kNN)
* Phan loai kNN bao gom hai giai doan:

    + Giai doan testing, kNN phan loai moi hinh anh thu nghiem bang cach so sanh voi tat ca cac hinh anh dao tao va chuyen cac nhan cua k vi du dao tao tuong tu nhat.
    + Gia tri k duoc xac thuc cheo (cross-validated)

```python
import random
import numpy as np
from cs231n.data_utils import load_CIFAR10
import matplotlib.pyplot as plt

#Ham nay de cho ket qua chay xuat hien ngay tren notebook nay
# chu khong phai chay tren 1 new window.
%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) #Set default sizze of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

#Ham nay de notebook se tai lai cac modules python;
#Xem tai: http://stackoverflow.com/questions/1907993/autoreload-of-modules-in-ipython
%load_ext autoreload
%autoreload 2
```

```python
#Load the raw CIFAR-10 data.
cifar10_dir = 'cs231n/datasets/cifar-10-batches-py/'
X_train, y_train, X_test, y_test = load_CIFAR10(cifar10_dir)

#De kiem tra tinh chinh xac, in ra kich thuoc cua du lieu dao tao va kiem tra
print 'Khung du lieu Training: ', X_train.shape
print 'Khung nhan Traning: ', y_train.shape
print 'Khung du lieu Test: ', X_test.shape
print 'Khung nhan Test', y_test.shape
```

```python
# Mot so vi du tu Dataset
# Hien thi mot vai vi du cua hinh anh Traing tu moi lop.
classes = ['May bay', 'Oto', 'Chim', 'Meo', 'Nai', 'cho', 'Ech', 'Ngua', 'Tau', 'Xe Tai']
num_classes = len(classes)
samples_per_class = 6

for y, cls in enumerate(classes): # y la chi so, cls = 'May bay'; enumerate cho phep lap va co 1 bo dem tu dong
    idxs = np.flatnonzero(y_train == y) 
    idxs = np.random.choice(idxs, samples_per_class, replace = False)
    for i, idx in enumerate(idxs):
        plt_idx = i * num_classes + y + 1
        plt.subplot(samples_per_class, num_classes, plt_idx)
        plt.imshow(X_train[idx].astype('uint8'))
        plt.axis('off')
        if i == 0:
            plt.title(cls)

plt.show()
```

```python
#Lay mau du lieu
num_training = 5000
mask = list(range(num_training))
X_train = X_train[mask]
y_train = y_train[mask]

num_test = 500
mask = list(range(num_test))
X_test = X_test[mask]
y_test = y_test[mask]
```

```python
#Dua du lieu hinh anh thanh cac hang
X_train = np.reshape(X_train, (X_train.shape[0], -1))
X_test = np.reshape(X_test, (X_test.shape[0], -1))
print(X_train.shape, X_test.shape)
```

```python
from cs231n.classifiers import KNearestNeighbor

#Tao 1 nhan kNN
#Viec training 1 nhan kNN la 1 noop:
# Nhan chi ghi nho du lieu va khong xu ly them nua
classifier = KNearestNeighbor()
classifier.train(X_train, y_train)
```

    Phan loai du lieu thu nghiem voi bo phan loai kNN. Co the pha vo qua trinh nay tu hai buoc:
    1. Tinh toan khoang cach giua tat ca cac vi du kiem tra va tat cac vi du train
    2. Voi khoang cach, doi voi moi vu du kiem tra, tim thay k vi du gan nhat va yeu cau gan nhan


    Cho phep bat dau voi viec tinh toan ma tran khoang cach giua tat ca cac vi du ve train va test. Vi du, neu co cac vi du train Ntr va cac vi du test Nte, giai doan nay se tao ra mot ma tran (Nte x Ntr) trong do moi phan tu (i,j) la khoang cac giau test thu i va train thu j.
    
    Dau tien, mo cs231n/classifiers/k_nearest_neighbor.py va trien khai ham compute_distances_two_loops su dung vong lap doi (khong hieu qua) tren tat ca cac cap (test, train) cac vi du va tinh toan ma tran khoang cach mot phan tu tai mot thoi diem

```python
%%time
#Mo cs231n/classifiers/k_nearest_neighbor.py va trien khai
#compute_distances_two_loops.

#Kiem tra qua trinh thuc hien:

dists = classifier.compute_distances_two_loops(X_test)
print (dists.shape)
```

```python
%%time
#Tang toc do tinh toan ma tran khoang cach bang cach su dung vector 1 loop
#Thuc hien ham compute_distances_one_loop va run code:
dists_one = classifier.compute_distances_one_loop(X_test)
print(dists_one.shape)

#De dam bao vector chinh xac, thuc hien naive. Co nhieu cach de xem
#hai ma tran co giong nhau khong; mot trong nhung cach don gian la dung Frobenius norm
# Frobenius norm cua 2 ma tran la hinh vuong root cua tong binh phuong khac nhau cua tat ca cac phan tu
# Noi cach khac, dinh hinh lai ma tran vao vector va tinh toan khoang cach Euclide.
difference = np.linalg.norm(dists - dists_one, ord='fro')
print('Su khac nhau: %f' % (difference, ))
if difference < 0.001:
    print ('Good! Cac ma tran khoang cach giong nhau!')
else:
    print('No! Cac ma tran khoang cach khac nhau!')
```

```python
%%time

#Thuc hien ban vector hoa day du trong compute_distance_no_loops
dists_two = classifier.compute_distances_no_loops(X_test)

#Kiem tra ma tran khoang cach co giong voi ma tran khoang cach da tinh truoc khong:
difference = np.linalg.norm(dists - dists_two, ord = 'fro')
print('Su khac nhau: %f' % (difference, ))
if difference < 0.001:
    print ('Good! Cac ma tran khoang cach giong nhau!')
else:
    print('No! Cac ma tran khoang cach khac nhau!')
```

```python
#So sanh toc do thuc hien
def time_function(f, *args):
    """
        Goi ham f voi arg va tra ve thoi gian (tinh bang giay) thuc hien
    """
    import time
    tic = time.time()
    f(*args)
    toc = time.time()
    return toc - tic

two_loop_time = time_function(classifier.compute_distances_two_loops, X_test)
print('Two loop version took %f seconds' % two_loop_time)

one_loop_time = time_function(classifier.compute_distances_one_loop, X_test)
print('One loop version took %f seconds' % one_loop_time)

no_loop_time = time_function(classifier.compute_distances_no_loops, X_test)
print('No loop version took %f seconds' % no_loop_time)


    
```

Cross-validation (Xac thuc cheo)
    Trien khai bo phan loai k-Nearest Neighbor, dat gai tri k = 5. Xac dinh gia tri tot nhat cua tham so nay voi Cross-validation

```python
num_folds = 5
k_choices = [1, 3, 5, 8, 10, 12, 15, 20, 50, 100]

X_train_folds = []
y_train_folds = []

#################################################################################
# TODO                                                                          #
# Chia nho du lieu train thanh cap folds (nep gap). Sau khi chia,               #
# X_train_folds va y_train_folds la danh sach co do dai num_folds,              #
# trong do, y_train_folds[i] la vector nhan cho cac diem trong X_train_folds[i] #
# Hint: dung ham array_split                                                    #
#################################################################################

X_train_folds = np.array_split(X_train, num_folds)
y_train_folds = np.array_split(y_train, num_folds)

#################################################################################
#                               END OF YOUR CODE                                #
#################################################################################

# 1 tu dien giu do chinh xac cho cac gia tri khac nhau cua k ma duoc tim thay
# khi xac thuc cheo. Sau khi chay xac thuc cheo,
# k_to_accuracies[k] phai la danh sach co do dai num_folds tao nen su khac nhau
# gia tri do chinh xac ma da tim thay khi su dung gia tri k do.
k_to_accuracies = {}

#################################################################################
# TODO                                                                          #
# Thuc hien xac nhan cheo k-fold de tim gia tri tot nhat cua k. Cho moi         #
# gia tri co the cua k, chay thuat toan k-nearest-neighbor voi num_folds lan    #
# trong moi truong hop su dung tat ca tu 1 trong cac nep gap lam data train va  #
# lan cuoi duoi dang tap xac thuc. Luu cac do chinh xac cho tat ca cac nep gap  #
# va tat cac gia tri cua k trong tu dien k_to_accuracies                        #
#################################################################################
for k in k_choices:
    k_accuracies = list()
    for i in range(num_folds):
        
        train_data = np.concatenate(X_train_folds[:i] + X_train_folds[i+1:])
        test_data = X_train_folds[i]
        
        train_labels = np.concatenate(y_train_folds[:i] + y_train_folds[i+1:])
        test_labels = y_train_folds[i]
        
        classifier = KNearestNeighbor()
        classifier.train(train_data, train_labels)
        dists_two = classifier.compute_distances_no_loops(test_data)
        y_test_pred = classifier.predict_labels(dists_two, k=k)
        num_correct = np.sum(y_test_pred == test_labels)
        num_test = test_labels.size
        accuracy = float(num_correct) / num_test
        k_accuracies.append(accuracy)
        
    k_to_accuracies[k] = k_accuracies


#################################################################################
#                               END OF YOUR CODE                                #
#################################################################################


#In ra cac do chinh xac duoc tinh toan
for k in sorted(k_to_accuracies):
    for accuracy in k_to_accuracies[k]:
        print ('k = %d, do chinh xac  = %f' % (k, accuracy))
```

```python
# plot the raw observations
for k in k_choices:
    accuracies = k_to_accuracies[k]
    plt.scatter([k] * len(accuracies), accuracies)
    
#Ve duong huong voi cac thanh loi tuong ung voi do lech chuan
accuracies_mean = np.array([np.mean(v) for k,v in sorted(k_to_accuracies.items())])
accuracies_std = np.array([np.std(v) for k,v in sorted(k_to_accuracies.items())])
plt.errorbar(k_choices, accuracies_mean, yerr=accuracies_std)
plt.title('Cross-validation (Xac thuc cheo) on k')
plt.xlabel('k')
plt.ylabel('Do chinh xac xac thuc cheo')
plt.show()
```

```python
#Dua tren ket qua xac thuc cheo o tren, hay chon gia tri tot nhat cho k,
# dao tao lai trinh phan loai bang cach su dung tat ca du lieu dao tao va kiem tra tren test data
# Dat duoc do chinh xac tren 20% tren du lieu test.
best_k = 10

classifier = KNearestNeighbor()
classifier.train(X_train, y_train)
y_test_pred = classifier.predict(X_test, k=best_k)

#Tinh toan va hien thi do chinh xac
num_correct = np.sum(y_test_pred == y_test)
num_test = 500
accuracy = float(num_correct) / num_test
print('Chinh xac: %d / %d => Do chinh xac: %f' % (num_correct, num_test, accuracy))
```
